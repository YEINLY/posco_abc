### 2-3. 컨테이너형 자료 (list, tuple, dict, set)

여러 값을 하나의 변수에 묶어서 담을 수 있는 자료형입니다.

1. 리스트(list)
   - 대괄호 `[]` 사용
   - 변경 가능 (mutable)
   - 예: `numbers = [1, 2, 3]`

2. 튜플(tuple)
   - 소괄호 `()` 사용
   - 변경 불가능 (immutable)
   - 예: `point = (10, 20)`

3. 딕셔너리(dict)
   - 중괄호 `{}` 사용
   - 키(key)와 값(value) 쌍으로 저장
   - 예: `person = {"name": "홍길동", "age": 30}`

4. 집합(set)
   - 중괄호 `{}` 사용
   - 중복을 허용하지 않음, 순서가 중요하지 않음
   - 예: `unique_numbers = {1, 2, 3}`

## 7. 반복문 (`for`, `while`, `range`, `break`)

### 7-1. for 문

반복 가능한 객체(리스트, 문자열, range 등)를 순회하며 반복합니다.

```python
for i in range(5):
    print(i)
```

### 7-2. range()

연속된 정수 범위를 만들어주는 함수입니다.

- `range(n)` : 0 ~ n-1
- `range(a, b)` : a ~ b-1
- `range(a, b, step)` : a부터 step씩 증가하며 b-1까지


````python
range(10) # 0 ~ 9
range(6, 10) # 6, 7, 8, 9
range(3, 13, 2) # 3, 5, 7, 9, 11
range(10, 0, -1) # 10, 9, 8, ..., 1
````

````python
for i in range(10, 0, -1):
    print(i)
````

````python
# for + range 예제

for i in range(5):
    print("i =", i)

print("---")

for i in range(2, 7):
    print("i =", i)

print("---")

for i in range(1, 10, 2):
    print("홀수 i =", i)
````

### 7-3. while 문

조건이 참인 동안 계속 반복합니다.

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

### 7-4. break 문

반복문을 중간에 강제로 종료합니다.


````python
i = 0
while i < 5:
    print(i)
    i += 1
````

````python
# while, break 예제

i = 1
while True:  # 무한 루프
    print("i =", i)
    if i >= 5:
        print("반복 종료")
        break  # 반복문 탈출
    i += 1
````

## 8. 함수와 메서드

### 8-1. 함수(Function)

특정 기능을 묶어 재사용할 수 있도록 한 코드 덩어리입니다.

```python
def 함수이름(매개변수1, 매개변수2):
    실행문
    return 반환값
```

### 8-2. 내장 함수 예시

- `print()`, `input()`, `max()`, `min()`, `sum()`, `len()` 등

### 8-3. 사용자 정의 함수

사용자가 직접 함수를 정의해서 사용할 수 있습니다.

### 8-4. 전역 변수와 지역 변수

- 전역 변수: 함수 밖에서 선언, 프로그램 전체에서 접근 가능
- 지역 변수: 함수 안에서 선언, 함수 안에서만 사용 가능

### 8-5. 메서드(Method)

객체(자료형, 예: 문자열, 리스트 등)에 속해 있는 함수입니다.

```python

s.upper()  # 문자열 메서드 예시
```


````python
# 내장 함수 예제

numbers = [10, 20, 30, 40]

print("최댓값:", max(numbers))
print("최솟값:", min(numbers))
print("합계:", sum(numbers))
print("개수:", len(numbers))
print(numbers)
````

````python
# 사용자 정의 함수 예제

def add(a, b = 10):
    # Docstring > 함수 설명 문구(코드 작동 X)
    """두 수 a, b를 더한 값을 반환하는 함수"""
    return a + b

result = add(3)
print("3 + 10 =", result)
````

## 9. 클래스와 객체

함수와 변수를 묶어서 하나의 설계도로 표현한 것을 클래스(class)라고 합니다.  
클래스를 이용하면 관련 있는 데이터(속성)와 동작(메서드)을 하나의 단위로 관리할 수 있습니다.

기본적인 클래스 정의 형태는 다음과 같습니다.

```python
class 클래스이름:
    def __init__(self, 매개변수1, 매개변수2):
        self.속성1 = 매개변수1
        self.속성2 = 매개변수2

    def 메서드이름(self):
        실행문
```

- `__init__` 메서드: 객체가 생성될 때 자동으로 호출되는 생성자
- `self`: 생성된 객체 자신을 가리키는 예약어 관례 (임의의 이름을 쓸 수도 있으나 일반적으로 self 사용)


````python
# 클래스 정의와 사용 예제
# 설계도 만들기
class Person:
    def __init__(self, x, y): # 초기화 함수를 통해서 클래스의 기본 속성을 정의
        self.name = x
        self.age = y

    def introduce(self): # 이 클래스만 독점적으로 사용할 수 있는 함수(method) 정의
        print(f"안녕하세요, 제 이름은 {self.name}이고, 나이는 {self.age}살입니다.")

# 객체(인스턴스) 생성
p1 = Person("홍길동", 30)
p2 = Person("김철수", 25)
````

````python
#속성 출력
print(p1.name)
````

````python
#속성 출력
print(p2.age)
````

````python
#메서드 사용
p1.introduce()
````

````python
#메서드 사용
p2.introduce()
````

````python
# 클래스 예제 2: 간단한 은행 계좌 클래스

class BankAccount:
    def __init__(self, x, y=0):
        self.owner = x
        self.balance = y

    def deposit(self, amount):
        self.balance += amount # balance = amount + balance
        print(amount, "원이 입금되었습니다. 현재 잔액:", self.balance)

    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        else:
            self.balance -= amount
            print(amount, "원이 출금되었습니다. 현재 잔액:", self.balance)

# 사용 예시
account = BankAccount("홍길동", 10000)
````

````python
account.balance
````

````python
account.deposit(5000000)
````

````python
account.withdraw(50000)
````

````python
# 직접 해보기
# TODO: 책 정보를 저장하는 Book 클래스를 정의해 보세요.
# 요구사항:
# - 속성: 제목(title), 저자(author), 페이지 수(pages)
# - 메서드: info() → "제목: ..., 저자: ..., 페이지: ...쪽" 형태로 출력

class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def info(self):
        print(f"제목: {self.title}, 저자: {self.author}, 페이지: {self.pages}쪽")

# 예시 실행 코드
my_book = Book("파이썬 입문", "홍길동", 300)
print(type(my_book))
````

````python
# 빈칸 채우기 문제 9
# 아래 코드의 ___ 부분을 채워서, Car 클래스를 완성해 보세요.
# Car 객체는 브랜드와 현재 속도를 속성으로 가지며,
# accelerate 메서드는 주어진 만큼 속도를 증가시킵니다.

class Car:
    def __init__(self, brand, speed=0):
        self.brand = brand
        self.speed = speed

    def accelerate(self, amount):
        self.speed += amount
        print("현재 속도:", self.speed)

my_car = Car("MyBrand")
type(my_car)
````

````python
my_car.accelerate(10)   # 현재 속도: 10
````

````python
my_car.accelerate(20)   # 현재 속도: 30
````
