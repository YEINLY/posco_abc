### 문제 5-1

```python
a = 10
b = 3
c = 10

print(a == c)           # True  (10 == 10)
print(a != b)           # True  (10 != 3)
print(b > 5 or a < 20)  # True  (False or True)
print(a > 5 and b > 5)  # False (True and False)
print(not (a == b))     # True  (not False)
```

### 문제 7-3 (while + break)

```python
i = 1
total = 0

while True:
    total += i
    if total > 50:
        break
    i += 1

print(f"숫자: {i}, 누적 합: {total}")
```

## 섹션 8. 함수와 메서드

### 문제 8-1

```python
def my_max(a, b):
    if a >= b:
        return a
    else:
        return b

print(my_max(3, 7))   # 7
print(my_max(10, 2))  # 10
```

### 문제 8-2

```python
def celsius_to_fahrenheit(c):
    return c * 9 / 5 + 32

print(celsius_to_fahrenheit(0))    # 32.0
print(celsius_to_fahrenheit(100))  # 212.0
```

### 문제 8-3 (문자열 메서드)

```python
s = "hello python, welcome to python world"

# 1) 대문자
print(s.upper())

# 2) 단어 리스트
print(s.split())

# 3) "python" 등장 횟수
print(s.count("python"))
```

### 문제 8-4 (빈칸 채우기)

```python
def my_min(lst):
    result = lst[0]
    for num in lst:
        if num < result:
            result = num
    return result

print(my_min([5, 2, 8, 1, 9]))  # 1
```

## 섹션 9. 클래스와 객체

### 문제 9

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def grade(self):
        if self.score >= 90:
            return "A"
        elif self.score >= 80:
            return "B"
        elif self.score >= 70:
            return "C"
        elif self.score >= 60:
            return "D"
        else:
            return "F"

s1 = Student("조예찬", 85)
s2 = Student("홍길동", 55)

print(f"{s1.name}: {s1.grade()}")  # 조예찬: B
print(f"{s2.name}: {s2.grade()}")  # 홍길동: F
```
